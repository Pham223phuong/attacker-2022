```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
```

```python
data = pd.read_excel("C:\\Users\\as\\Downloads\\FINAL_train_attacker.xlsx")
print(data)
```

```python
data['cat_12'].fillna(value=data['cat_12'].mode()[0], inplace=True)
data['cat_9'].fillna(value=data['cat_9'].mode()[0], inplace=True)
data['location_id'].fillna(value=data['location_id'].mode()[0], inplace=True)
data.to_excel("C:\\Users\\as\\Downloads\\FINAL_train_attacker.xlsx")
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
f, ax = plt.subplots(figsize=(15, 6))
plt.xticks(rotation='90')
sns.barplot(x=missing_data.index, y=missing_data['Percent'])
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
missing_data.head()
```
